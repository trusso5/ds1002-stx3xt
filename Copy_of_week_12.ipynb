## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(slices, labels=labels)
plt.show()
```
